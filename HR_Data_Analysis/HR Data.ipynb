# PANDAS ANALYSIS OF HR DATA

```python
# Importing required libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.ticker as mticker
```

```python
# Import data
hr_data = pd.read_csv(r"HR DATA\HR_Data_MNC.csv")
```

```python
# Copy data in new dataframe
df = hr_data.copy()
```

```python
# Checking random data
df.sample(n=10)
```

```python
# Basic info about data
df.info()
```

```python
# Drop unwanted column
df = df.drop("Unnamed: 0", axis = 1)
```

```python
df
```

```python
# Change data type for Hire_Date
df["Hire_Date"]= pd.to_datetime(df["Hire_Date"])
```

```python
# Ratings
df["Performance_Rating"].value_counts()
```

```python
# Average rating Employees received
df["Performance_Rating"].mean()
```

```python
# Average Experience
df["Experience_Years"].mean()
```

```python
# Employees by experience
df["Experience_Years"].value_counts()

plt.figure(figsize=(10,6)) 
sns.set(style="whitegrid", palette="Set2") 
order = df["Experience_Years"].value_counts().index  
ax = sns.countplot(x="Experience_Years", data=df, order=order, palette="pastel", hue = "Experience_Years", legend = False)
plt.title("Employee Distribution by Experience (Years)", fontsize=14, fontweight="bold")
plt.xlabel("Years of Experience", fontsize=12)
plt.ylabel("Number of Employees", fontsize=12)
for container in ax.containers:
    ax.bar_label(container, fmt='%d', label_type='edge', fontsize=10)
plt.tight_layout()
plt.show()
```

```python
# Employees experience by percentage
plt.figure(figsize=(10,6))
order = df["Experience_Years"].value_counts().index
ax = sns.countplot(x="Experience_Years", data=df, order=order)
total = len(df)
for container in ax.containers:
    ax.bar_label(container, labels=[f'{(v/total*100):.2f}%' for v in container.datavalues], label_type='edge')

plt.title("Experience Distribution (%)", fontsize=14, fontweight="bold")
plt.xlabel("Years of Experience")
plt.ylabel("Number of Employees")
plt.tight_layout()
plt.show()
```

```python
# Distribution of Employee status
status = df["Status"].value_counts()
plt.figure(figsize=(6, 6))
plt.pie(
    status,
    labels=status.index,
    autopct='%1.1f%%',
    startangle=90,
    counterclock=False,
    wedgeprops={'edgecolor': 'white'},
    explode = (0.03,0.03,0.03,0.03)
)
plt.title("Distribution of Employee Status", fontsize=14)
plt.tight_layout()
plt.show()
```

```python
# Distribution of Work modes
work_mode = df["Work_Mode"].value_counts()
```

```python
counts = df["Work_Mode"].value_counts() / 1e6

plt.figure(figsize=(6, 8))
sns.set(style="whitegrid", palette="pastel")
ax = sns.barplot(
    x=counts.index,
    y=counts.values,
    palette="pastel",
    hue=counts.index, 
    legend=False 
)
for i, v in enumerate(counts.values):
    ax.text(
        i,                     
        v + 0.02,              
        f"{v:.2f}M",          
        ha='center',       
        fontsize=10,
        fontweight='bold'
    )
plt.title("Employee Work Mode Distribution", fontsize=14, fontweight="bold")
plt.xlabel("Work Mode", fontsize=12)
plt.ylabel("Employees (Millions)", fontsize=12)
plt.tight_layout()
plt.show()
```

```python
# Distribution of Employees in each department
department_employees = df["Department"].value_counts()
plt.figure(figsize=(10, 6)) 
ax = sns.barplot(
    x=department_employees.values, 
    y=department_employees.index,   
    palette="plasma",
    hue=department_employees.index,
    legend=False
)
plt.title("Distribution of Employees by Department", fontsize=16)
plt.xlabel('Number of Employees', fontsize=12)
plt.ylabel('Departments', fontsize=12)
plt.show()
```

```python
# Distribution of Employees by Job Title
department_employees = df["Job_Title"].value_counts()
plt.figure(figsize=(10, 6)) 
ax = sns.barplot(
    x=department_employees.values, 
    y=department_employees.index,   
    palette="plasma",
    hue=department_employees.index,
    legend=False
)
plt.title("Distribution of Employees by Job Title", fontsize=16)
plt.xlabel('Number of Employees', fontsize=12)
plt.ylabel('Designation', fontsize=12)
plt.show()
```

```python
# Global setting for float
pd.options.display.float_format = '{:,.2f}'.format

# Average salary by department
average_salary = df.groupby("Department")["Salary_INR"].mean().round(2)
average_salary_lakhs = average_salary / 1e5

plt.figure(figsize=(10, 8))
ax = sns.barplot(
    x=average_salary_lakhs.index,
    y=average_salary_lakhs.values,
    palette="plasma",
    hue=average_salary_lakhs.index,
    legend=False
)

for i, v in enumerate(average_salary_lakhs.values):
    ax.text(
        i,
        v + 0.5,  
        f"{v:.2f} L",
        ha='center',
        fontsize=10,
        fontweight='bold'
    )
plt.title("Average Salary by Department", fontsize=14, fontweight='bold')
plt.xlabel("Department", fontsize=12)
plt.ylabel("Average Salary (Lakhs)", fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# Job title with highest average salary
df['Salary_Lakhs'] = df['Salary_INR'] / 100000
average_salary_title = df.groupby("Job_Title")["Salary_Lakhs"].mean().sort_values(ascending=False)

plt.figure(figsize=(10, 8)) 
ax = sns.barplot(
    x=average_salary_title.values,
    y=average_salary_title.index,
    palette="plasma",
    hue = average_salary_title.index
)

for i, v in enumerate(average_salary_title.values):
    ax.text(
        v + 0.5,  
        i,        
        f"{v:.2f} L",
        ha='left', 
        va='center', 
        fontsize=10,
        fontweight='bold'
    )
plt.title("Average Salary by Job Title", fontsize=16, fontweight='bold')
plt.xlabel("Average Salary (Lakhs INR)", fontsize=12)
plt.ylabel("Job Title", fontsize=12) 
plt.tight_layout()
plt.show()
```

```python
# Number of employees resigned and terminated in each department
df_resigned = df[df["Status"]=="Resigned"]
print(df_resigned)
```

```python
resigned_emp=df_resigned.groupby("Department")["Status"].count()
```

```python
plt.figure(figsize=(10,8))
sns.barplot(x=resigned_emp.index, y=resigned_emp.values, palette="plasma", hue=resigned_emp.index)
plt.title("Number of employees Resigned from each department")
plt.xlabel("Department")
plt.ylabel("Number of Employees")
plt.show()
```

```python
df_terminated = df[df["Status"]=="Terminated"]
terminated_emp=df_terminated.groupby("Department")["Status"].count()
plt.figure(figsize=(10,8))
sns.barplot(x=terminated_emp.index, y=terminated_emp.values, palette="plasma", hue=terminated_emp.index)
plt.title("Number of employees terminated from each department")
plt.xlabel("Department")
plt.ylabel("Number of Employees")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

filtered = df[df["Status"].isin(["Resigned", "Terminated"])]

plt.figure(figsize=(10, 8))
sns.countplot(data=filtered, x="Department", hue="Status", palette="plasma")
plt.title("Resigned vs Terminated Employees by Department", fontsize=14)
plt.xlabel("Department")
plt.ylabel("Number of Employees")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
#Number of Employees hired each year
df["Year"]=df["Hire_Date"].dt.year
year_hire=df.groupby("Year")["Employee_ID"].count()
plt.figure(figsize=(10,8))
sns.barplot(x=year_hire.index, y=year_hire.values, palette="plasma", hue=year_hire.index, legend=False)
plt.title("Number of employees Hired each yaer")
plt.xlabel("Year")
plt.ylabel("Number of Employees")
plt.show()
```

```python
#Department with highest attrition rate
dept_cnt=df.groupby("Department")["Status"].agg(total_emp="count",resigned=lambda x:(x=="Resigned").sum())
```
