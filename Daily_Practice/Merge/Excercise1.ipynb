```python
import pandas as pd
import numpy as np
```

```python
url1 = "https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/05_Merge/Auto_MPG/cars1.csv"
url2 = "https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/05_Merge/Auto_MPG/cars2.csv"
```

```python
df1 = pd.read_csv(url1)
df2 = pd.read_csv(url2)
```

```python
print(df1.head())
print(df2.head())
```

### first dataset has some unnamed blank columns, fix df1

```python
df1 = df1.loc[: , "mpg":"car"]
df1.head()
```

### What is the number of observations in each dataset?

```python
print(df1.shape)
print(df2.shape)
```

Join df1 and df2 into a single DataFrame called cars

```python
cars = pd.concat([df1,df2])
cars.head()
```

### there is a column missing, called owners. Create a random number Series from 15,000 to 73,000.

```python
car_owners = np.random.randint(1500, high=73001, size = 398, dtype=np.int32)
car_owners
```

### Add the column owners to cars

```python
cars["car_owners"] = car_owners
cars.head()
```
