```python
import pandas as pd
import numpy as np
```

###  Create 3 differents Series, each of length 100, as follows:
- The first a random number from 1 to 4
- The second a random number from 1 to 3
- The third a random number from 10,000 to 30,000

```python
s1 = pd.Series(np.random.randint(1, high = 5, size = 100, dtype = np.int32))
s2 = pd.Series(np.random.randint(1, high = 3, size = 100, dtype = np.int32))
s3 = pd.Series(np.random.randint(10000, high = 30000, size = 100, dtype = np.int32))

print(s1, s2, s3)
```

### Let's create a DataFrame by joinning the Series by column

```python
dataset = pd.concat([s1, s2, s3], axis =1)
dataset.head()
```

### Change the name of the columns to bedrs, bathrs, price_sqr_meter

```python
dataset.rename(columns = {0: ' bedrs', 1: 'bathrs' , 2:'price_sqr_meter'}, inplace=True)
dataset.head()
```

### Create a one column DataFrame with the values of the 3 Series and assign it to 'bigcolumn'

```python
bigcolumn = pd.concat([s1, s2, s3], axis=0)
bigcolumn = bigcolumn.to_frame()
print(type(bigcolumn))
bigcolumn
```
