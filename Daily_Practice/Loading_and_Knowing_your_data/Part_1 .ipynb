### Most ordered Item

```python
most_ordered = df.groupby('item_name')['quantity'].sum().sort_values(ascending=False)
```

```python
most_ordered.head(1)
```

### Which item was included in the most orders

```python
frequency = df['item_name'].value_counts()
frequency.head(1)
```

### What was the most ordered item in the choice_description column?

```python
most_ordered = df.groupby('choice_description')['quantity'].sum().sort_values(ascending=False)
most_ordered.head(1)
```

### How many items were orderd in total?

```python
total_item = df.quantity.sum()
total_item
```

### Turn the item price into a float

```python
df.item_price.dtype
```

```python
df['item_price'] = df.item_price.str.replace('$', '', regex=False).str.strip().astype(float)
```

```python
df.item_price.dtype
```

### How much was the revenue for the period in the dataset?

```python
revenue = (df['quantity'] * df['item_price']).sum()
revenue
```

### How many orders were made in the period?

```python
df.order_id.value_counts().count()
```

### What is the average revenue amount per order?

```python
revenue = (df['quantity'] * df['item_price']).sum()
order_count = df.order_id.nunique() 
average_revenue_per_order = revenue / order_count

average_revenue_per_order
```
