## **Loading and Knowing your data**

```python
import pandas as pd
import numpy as np
```

```python
url = 'https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv'
df = pd.read_csv(url, sep = '\t')
```

```python
df.head()
```

### How many products cost more than $10.00?

```python
df['item_price'] = df['item_price'].str.replace('$','',regex=False).astype(float)
```

```python
df['price_per_item'] = df['item_price'] / df['quantity']
df.head()
```

```python
df_pro = (df['price_per_item'] > 10).sum()
df_pro
```

###  What is the price of each item?

```python
item_prices = df[['item_name', 'choice_description', 'price_per_item']].drop_duplicates()
item_prices
```

### Find the most expensive items ?

```python
item_prices.sort_values(by = 'price_per_item', ascending = False)
```

### How many times was a Veggie Salad Bowl ordered?

```python
df[df['item_name'] == 'Veggie Salad Bowl']['quantity'].sum()
```
