```python
df_country = pd.read_excel("C:/Users/avina/Downloads/New folder (6)/Country-Code.xlsx")
```

```python
df_country.head()
```

```python
df_final = pd.merge(left=df, right=df_country,on = "Country Code", how="left")
df_final.head()
```

```python
# To get the count of value inside country column
df_final.Country.value_counts()
```

```python
country_names = df_final.Country.value_counts().index
```

```python
country_val = df_final.Country.value_counts().values
```

```python
# Top 3 countries
plt.pie(country_val[:3],labels = country_names[:3], autopct="%1.2f%%")
```

```python
# Grouping data based on ratings
df_final.groupby(["Aggregate rating", "Rating color","Rating text"]).size()
```

```python
ratings = df_final.groupby(["Aggregate rating", "Rating color","Rating text"]).size().reset_index().rename(columns={0:"Rating Count"})
```

```python
ratings
```

```python
matplotlib.rcParams["figure.figsize"]=(12,6)
sns.barplot(x = "Aggregate rating", y="Rating Count", data=ratings)
```

```python
sns.barplot(x = "Aggregate rating", y="Rating Count", data=ratings, hue="Rating color", palette = ["Brown", "red", "orange", "yellow", "green", ])
```

```python
# Based on count of range 
sns.countplot(x="Rating color", data = ratings, palette = ["Brown", "red", "orange", "yellow", "green", ])
```
